Modificando o nome das categorias:

```python
senioridade = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'Integral',
    'PT': 'Parcial',
    'CT': 'Contrato',
    'FL': 'Freelance'
}

df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
tamanho_empresa = {
    'S': 'Grande',
    'M': 'Média',
    'L': 'Pequena'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
mapa_trabalho = {
    0 : 'Presencial',
    100 : 'Remoto',
    50 : 'Hibrido'
}

df['remoto'] = df['remoto'].replace(mapa_trabalho)
df['remoto'].value_counts()
```

Exemplos fictícios

Exemplo de preenchimento com média e mediana

```python
import numpy as np
# Criação de um Data Frame
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})
# calcula a média e substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

''' Calcula a mediana e substitui os nulos pela mediana'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

Exemplo de preenchimento com o valor anterior

Usar ffill (forward fill)

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

# completando valores nulos com valores anteriores
df_temperaturas['Preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

Exemplo de preenchimento com o valor posterior

Usar bfill (backward fill)

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

# completando valores nulos com valores posteriores
df_temperaturas['Preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

Exemplo de preenchimento com valor fixo

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidades': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']
})

df_cidades['cidade_preenchida'] = df_cidades['cidades'].fillna('Não informado')
df_cidades
```

```python
import plotly.express as px
```

```python
# prompt: Crie um gráfico de média salarial por senioridade em barras usando o plotly

senioridade_media_salario = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()
fig = px.bar(senioridade_media_salario,
             x='senioridade',
             y='usd',
             title='Média Salarial por Senioridade',
             labels={'senioridade': 'Nivel de Senioridade', 'usd': 'Média Salarial Anual (USD)'})
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
            names='tipo_trabalho',
            values='quantidade',
            title='Proporão dos tipos de trabalho')

fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
            names='tipo_trabalho',
            values='quantidade',
            title='Proporão dos tipos de trabalho',
             hole=0.5)

fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
            names='tipo_trabalho',
            values='quantidade',
            title='Proporão dos tipos de trabalho',
             hole=0.5)

fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
# Filtrar o DataFrame para incluir apenas 'Data Scientist'
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist'].copy()

# Calcular a média salarial por país para Data Scientists
media_salario_pais_ds = df_ds.groupby('residencia')['usd'].mean().sort_values(ascending=False).reset_index()

# Criar o gráfico de barras interativo com Plotly
fig = px.bar(media_salario_pais_ds,
              x='residencia',
              y='usd',
              title='Média Salarial de Data Scientists por País (USD)',
              labels={'residencia': 'País de Residência', 'usd': 'Média Salarial Anual (USD)'})

# Mostrar o gráfico
fig.show()
```
